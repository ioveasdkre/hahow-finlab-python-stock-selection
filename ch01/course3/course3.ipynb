```python
date = pd.date_range('20180101', periods=6)
s = pd.Series([1, 2, 3, 4, 5, 6], index=date)
s
```

每個時間都對應到一個值，這就是傳說中的time series了！

接下來，我們也可以直接創造全部都是0的time series喔！只要把list換成一個任意數字就可以了：


```python
s = pd.Series(2, index=date)
s
```

## 查找


```python
# 建立一個 time series
date = pd.date_range('20180101', periods=6)
s = pd.Series([1, 2, 3, 4, 5, 6], index=date)

# 查找
s.loc['20180104']
```

```python
s.loc['20180102':'2018-01-04']
```

```python
s.iloc[1]
```

```python
s.iloc[1:4]
```

以上四種方式，大家搞清楚了嗎？

既然可以查找，那接下來，我們就來修改一下序列吧！

## 修改


```python
date_value = [1, 2, 3, 4, 5, 6]
date_index = pd.date_range('20180101', periods=6)

s = pd.Series(date_value, index=date_index)
s
```

```python
print(s)
# s.max()
# s.min()
# s.mean() # 平均值
s.std()
```

```python
print(s)
# s.cumsum()  # 不斷累加
# s.cumprod() # 不斷相乘
```

```python
print(s)
s.rolling(2).sum()  # 連續兩筆 相加
# s.rolling(2).max() # 連續兩筆 最大值
# s.rolling(2).min()
# s.rolling(2).mean()
# s.rolling(2).std()
```

```python
print(s)
# s + 1
# s - 1
# s * 2
# s / 2
# s > 3
s < 3
```

## 繪圖


```python
%matplotlib inline
s.plot()
```

## 綜合應用！


```python
larger_than_3 = s > 3
s.loc[larger_than_3]
```

```python
s.loc[larger_than_3] = s.loc[larger_than_3] + 1
s
```

接下來你會看到一串可怕的運算，但別擔心，我們把它拆解開來看！

- 從頭開始讀
- 遇到會先算


```python
(s.rolling(2).sum().cumsum() + 1).max()
```

```python
# 在跑出它之前，先用手算算看
# 究竟會跑出多少呢？
((s + 2).cumprod() + 4).min()
```

# DataFrame 介紹！


DataFrame 基本上，可以想像是很多的time series，

我們剛剛所學，都是修改其中一條 Series，但我們藉由dataframe，就能以相同的function，直接修改多個Series喔！


```python
s1 = pd.Series([1, 2, 3, 4, 5, 6], index=date)
s2 = pd.Series([5, 6, 7, 8, 9, 10], index=date)
s3 = pd.Series([11, 12, 5, 7, 8, 2], index=date)

dictionary = {
    'c1': s1,
    'c2': s2,
    'c3': s3,
}

df = pd.DataFrame(dictionary)
df
```

```python
%matplotlib inline
df.plot()
```

## 選取

我們選取方式，幾乎跟剛剛一模一樣：


```python
df.loc['2018-01-02']
```

```python
df.iloc[1]
```

```python
df.loc['2018-01-02':'2018-01-05', ['c1', 'c2']]
```

```python
df.iloc[1:4, [0, 1]]
```

```python
df.cumsum()  # 累加
```

```python
df.cumprod()
```

```python
df.rolling(2).mean()
```

大家會發現，就是每條序列都執行一次原本series的功能而已！所以dataframe就這麼快速的教完囉！
