### 網頁爬蟲

今天我們先教最簡單的一種，之後我們會碰到各種不同的情況，
今天這招幾乎可以破解60%以上的網頁

首先我們要先import requests，可以把它想像成是瀏覽器


```python
import requests

res = requests.get('https://www.w3schools.com/html/html_tables.asp')
```

### 網頁代碼

取得網頁的原始碼


```python
res.text
```

### 轉碼


```python
res.encoding = 'utf-8'  # 'big5'
res.text
```

### 轉成DataFrame

接下來的步驟，我們要將文字檔中的table轉成DataFrame，主要可以分成兩個步驟：

- 存成 html 檔
- 用 pandas 讀取 html 檔


```python
# 存檔

f = open('test.html', 'w', encoding='utf-8')
f.write(res.text)
f.close()
```

```python
# 讀檔並轉成DataFrame

import pandas as pd

dfs = pd.read_html('test.html')
```

```python
# 瀏覽
dfs[0]
```

### 轉成DataFrame（快速版）

其實也可以只寫成一個步驟（但還是建議先存起來）


```python
# 不存檔直接轉成DataFrame
from io import StringIO

dfs = pd.read_html(StringIO(res.text))
dfs[0]
```
